```python
insert_query = """
INSERT INTO police_logs (
    stop_date, stop_time, country_name, driver_gender, driver_age, driver_race,
    violation, search_conducted, search_type, stop_outcome, drugs_related_stop,
    stop_duration, vehicle_number, timestamp
) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
"""
conn.commit()
```

```python
for _, row in data.iterrows():
    cursor.execute(insert_query, (
        row['stop_date'] if pd.notna(row['stop_date']) else None,
        row['stop_time'] if pd.notna(row['stop_time']) else None,
        row.get('country_name'),
        row.get('driver_gender'),
        int(row.get('driver_age')) if pd.notna(row.get('driver_age')) else None,
        row.get('driver_race'),
        row.get('violation'),
        row.get('search_conducted'),
        row.get('search_type'),
        row.get('stop_outcome'),
        row['drugs_related_stop'] if pd.notna(row['drugs_related_stop']) else None,
        row.get('stop_duration'),
        row.get('vehicle_number') if pd.notna(row.get('vehicle_number')) else f'VEH{row.name}',
        row.get('timestamp') if pd.notna(row.get('timestamp')) else None,
    ))
```
